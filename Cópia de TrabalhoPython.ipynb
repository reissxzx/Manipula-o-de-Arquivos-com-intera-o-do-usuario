```python
#FUNÇÃO 1 - LISTA DE ESCOLHA

from google.colab import files
import os

def ListaEscolha():

# Obtém a lista de arquivos no diretório atual (normalmente /content/ no Colab)
# Filtra para incluir apenas arquivos (excluindo diretórios)
  arquivosDisponiveis = [f for f in os.listdir('.') if os.path.isfile(f)]

  if not arquivosDisponiveis:
    print("Não há arquivos disponíveis para acesso no diretório atual.")
  else:
    print("Arquivos disponíveis para acesso:")
    # Mostra a lista de arquivos numerada para o usuário
    for i, arquivo in enumerate(arquivosDisponiveis):
        print(f"{i + 1}: {arquivo}")

    try:
        # Solicita ao usuário que escolha um arquivo
        escolha = input(f"Digite o número do arquivo que deseja acessar (1 a {len(arquivosDisponiveis)}): ")
        nomeArquivo = int(escolha) - 1 # Converte a escolha para índice baseado em 0

        # Verifica se o índice escolhido é válido
        if 0 <= nomeArquivo < len(arquivosDisponiveis):
            nomeArquivo = arquivosDisponiveis[nomeArquivo]

    except ValueError:
        print("Entrada inválida. Por favor, digite um número.")
    except Exception as e:
        print(f"Ocorreu um erro inesperado durante o processo de seleção: {e}")

    return nomeArquivo
```

```python
# FUNÇÃO 2 - OPÇÕES DO QUE FAZER

def seraFeito():
  import os
  from google.colab import files

  if nomeArquivo != None:
    oQueFazer = input('O que você quer fazer? 1-Escrever 2-Ler 3-Adicionar 4-Excluir 5-Baixar:  ')

    #Para Escrever
    if oQueFazer == '1':
      with open(nomeArquivo,'w',encoding='utf-8') as arquivo:
        arquivo.write(input('Digite o conteúdo do arquivo: '))
        arquivo.write('\n')
        print()
        print(f'O conteúdo foi adicionado ao arquivo {nomeArquivo}')
    #Para Ler
    elif oQueFazer == '2':
      with open(nomeArquivo,'r', encoding='utf-8') as arquivo:
        conteudo = arquivo.read()
        print(f'Conteúdo de {nomeArquivo}')
        print()
        print(conteudo)
    #Para Adicionar
    elif oQueFazer == '3':
      with open(nomeArquivo,'a',encoding='utf-8') as arquivo:
        arquivo.write(input('Digite o conteúdo do arquivo: '))
        print()
        print(f'O conteúdo foi adicionado ao arquivo {nomeArquivo}')
    #Para Excluir
    elif oQueFazer == '4':
      ArquivoParaExcluir = nomeArquivo
      try:
        os.remove(ArquivoParaExcluir)
        print(f"Arquivo '{ArquivoParaExcluir}' excluído com sucesso.")
      except FileNotFoundError:
        print(f"Arquivo '{ArquivoParaExcluir}' não encontrado.")
      except Exception as e:
        print(f"Ocorreu um erro ao excluir o arquivo: {e}")
    #Para Baixar
    elif oQueFazer=='5':
      ArquivoParaBaixar = nomeArquivo
      try:
        files.download(ArquivoParaBaixar)
        print(f"O arquivo '{nomeArquivo}' foi salvo com sucesso!")
      except FileNotFoundError:
        print(f"Erro: O arquivo '{nomeArquivo}' não existe!")
      except Exception as e:
        print(f'Ocorreu um erro durante o download: {e}')
    #Caso houver algum erro
    else:
      print('Nenhuma das alternativas foram selecionadas!')

    return
```

```python
# FUNÇÃO - CRIAR ARQUIVO

def CriarArquivo():
  with open(nomeArquivo,'w',encoding='utf-8') as arquivo:
    arquivo.write(input('Digite o conteúdo do arquivo: '))
    arquivo.write('\n')
    print()
    print(f'O conteúdo foi adicionado ao arquivo {nomeArquivo}')
  return
```
